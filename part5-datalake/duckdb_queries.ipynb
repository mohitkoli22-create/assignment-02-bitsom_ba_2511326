```python
#Q1 — Q1: List all customers along with the total number of orders they have placed
query = """
SELECT
    c.customer_id,
    c.name,
    COUNT(o.order_id) AS total_orders
FROM read_csv_auto('customers.csv') c
LEFT JOIN read_json_auto('orders.json') o
ON c.customer_id = o.customer_id
GROUP BY c.customer_id, c.name;
"""

duckdb.sql(query).df()
```

```python
# Q2 — Find the top 3 customers by total order value
query = """
SELECT
    c.customer_id,
    c.name,
    SUM(o.total_amount) AS total_spent
FROM read_csv_auto('customers.csv') c
JOIN read_json_auto('orders.json') o
ON c.customer_id = o.customer_id
GROUP BY c.customer_id, c.name
ORDER BY total_spent DESC
LIMIT 3;
"""

duckdb.sql(query).df()
```

```python
# Q3 — List all products purchased by customers from Bangalore

query = """
SELECT DISTINCT
    p.product_name
FROM read_csv_auto('customers.csv') c
JOIN read_json_auto('orders.json') o
ON c.customer_id = o.customer_id
JOIN read_parquet('products.parquet') p
ON o.product_id = p.product_id
WHERE c.city = 'Bangalore';
"""

duckdb.sql(query).df()
```

```python
# Q4  Join all three files to show: customer name, order date, product name, and quantity

query = """
SELECT
    c.name AS customer_name,
    o.order_date,
    p.product_name,
    o.quantity
FROM read_csv_auto('customers.csv') c
JOIN read_json_auto('orders.json') o
ON c.customer_id = o.customer_id
JOIN read_parquet('products.parquet') p
ON o.product_id = p.product_id;
"""

duckdb.sql(query).df()
```
